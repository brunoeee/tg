# Dados Semipreparados / Dados Preparados 1 Conv 1 Pooling

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2])
my_xticks = ['CNN 1','CNN 2','CNN 3']

# 1 conv
avg_ro_y = np.array([0.748415,0.781330,0.807990])
avg_sm_y = np.array([0.770579,0.786391,0.786011])
max_ro_y = np.array([0.733849,0.759128,0.784992])
max_sm_y = np.array([0.773711,0.784757,0.782168])

fig, axes = plt.subplots(ncols=2, figsize=(8, 4))
ax = axes[0]
ax.plot(x, avg_ro_y, '.-', label='AVG RO')
ax.plot(x, avg_sm_y, 'o-', label='AVG SM')
ax.plot(x, max_ro_y, 'D-', label='MAX RO')
ax.plot(x, max_sm_y, 's-', label='MAX SM')
ax.set_title("Tabela 6")
ax.set_xticks(x)
ax.set_xticklabels(my_xticks)
ax.grid(True)
ax.set_ylabel('AUC_ROC')
ax.legend(loc="lower right")

# 2 conv
avg_ro_y = np.array([0.743773,0.781345,0.802261])
avg_sm_y = np.array([0.774738,0.791013,0.792316])
max_ro_y = np.array([0.740326,0.765547,0.783584])
max_sm_y = np.array([0.766437,0.775704,0.789117])


ax = axes[1]
ax.plot(x, avg_ro_y, '.-', label='AVG RO')
ax.plot(x, avg_sm_y, 'o-', label='AVG SM')
ax.plot(x, max_ro_y, 'D-', label='MAX RO')
ax.plot(x, max_sm_y, 's-', label='MAX SM')
ax.set_title("Tabela 8")
ax.set_xticks(x)
ax.set_xticklabels(my_xticks)
ax.grid(True)
ax.legend(loc="lower right")

plt.show()
```
